## What is a Data Type?
A data type defines the type of value a variable can store. Python automatically assigns data types based on the value.

## 1. Integer (int)
Integers are whole numbers without decimal points.

```python
x = 10
y = -5
print(x, type(x))
print(y, type(y))
```

## 2. Floating Point (float)
Float stores numbers with decimal points.

```python
pi = 3.14
value = 2.5
print(pi, type(pi))
print(value, type(value))
```

## 3. String (str)
String is used to store text.

```python
name = 'Asad'
msg = 'Python is easy'
print(name, type(name))
print(msg, type(msg))
```

## 4. Boolean (bool)
Boolean represents True or False.

```python
a = True
b = False
print(a, type(a))
print(b, type(b))
```

## 5. List (list)
Ordered and changeable collection.

```python
nums = [1,2,3]
names = ['Ali','Sara']
print(nums, type(nums))
print(names, type(names))
```

## 6. Tuple (tuple)
Ordered but unchangeable collection.

```python
t = (10,20)
print(t, type(t))
```

## 7. Set (set)
Unordered collection of unique items.

```python
s = {1,2,2,3}
print(s, type(s))
```

## 8. Dictionary (dict)
Key-value pair data type.

```python
d = {'name':'Asad','age':25}
print(d, type(d))
```

## 9. Type Conversion

```python
x = 5
print(float(x))
print(str(x))
```

## ‚úÖ Summary
Python supports multiple built-in data types. Use type() to check them.
